```python
from transformers import pipeline
import random

# Use a small free text generation model (you can switch to OpenAI API if available)
quiz_gen = pipeline("text-generation", model="gpt2")

def generate_quiz(subject="Python", level="Beginner", difficulty="Easy"):
    prompt = f"Generate 10 multiple choice questions with 4 options each on {subject} at {level} level, difficulty {difficulty}. Each question should include the correct answer."

    quiz_text = quiz_gen(prompt, max_length=800, num_return_sequences=1)[0]['generated_text']

    # Simple formatting cleanup
    questions = []
    for line in quiz_text.split("\n"):
        if '?' in line:
            questions.append(line.strip())
    return questions[:10]

# Test the quiz generator
sample_quiz = generate_quiz("Python", "Beginner", "Easy")
for q in sample_quiz:
    print(q)
```
